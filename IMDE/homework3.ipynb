This method is applicable since we can create a surjection for our criteria. Let's assign the scale as follows:

1. Experience: 1-3 (>100, 100-200, >200)
2. Ease-of-Use: 1-3 (<4, 4-7, >7)
3. Documentation: 1-3 (<4, 4-7, >7)
4. Specificity: 1-3 (<6, 6-9, >9)
5. Lightweightness: 1-3 (<4, 4-7, >7)
6. Customizability: 1-3 (<4, 4-7, >7)

And the benchmarks as follows:

1. Experience: 2 (100-200)
2. Ease-of-Use: 3 (>7)
3. Documentation: 2 (4-7)
4. Specificity: 3 (>9)
5. Lightweightness: 3 (>7)
6. Customizability: 2 (4-7)

Now let's check what alternatives have reached what benchmarks and arrange them accordingly:

1. manim reaches all of them
2. pygame misses 2 of them
3. Unity misses 3 of them
4. Processing reaches only 1 benchmark
5. Pure Javascript reaches only 1 benchmark, but has lower average scores, so I will place it last

```python
def convert_value(value, benchmarks):
    if value < benchmarks[0]:
        return 1
    elif benchmarks[0] <= value <= benchmarks[1]:
        return 2
    else:  
        return 3

benchmarks = {
    'Experience': [2, 4],
    'Ease-of-Use': [4, 7],
    'Documentation': [4, 7],
    'Specificity': [6, 9],
    'Lightweightness': [4, 7],
    'Customizability': [4, 7]
}

for col in benchmarks:
    df[col] = df[col].apply(lambda x: convert_value(x, benchmarks[col]))

new_df = df[0:5].copy()
new_df.loc[len(new_df)] = {"Utility": "Benchmarks", "Experience": 2,
                   "Ease-of-Use": 3, "Documentation": 2,
                   "Specificity": 3, "Lightweightness": 3,
                   "Customizability": 2}
new_df
```

4. (1 балл) Метод целевого программирования (можете посчитать для всех трех метрик,
которые мы обсуждали на занятии, либо посчитать для одной из них, обосновав свой выбор)
[Подиновский, §6.1; Петровский, §8.7]. Можете использовать метрики с весами, веса можно
взять из SMART или амплитудные.

---

I will use Chebyshev's metric because it takes into account the maximal deviation from the optimum (perfect) value per each criteria and looks at the metric which is furthest away, which allows us to consider the worst, lacking metrics into account. Weights would be the same as we have used above. I will define the optimal point as the $(10, 10, 10, 10, 10, 10)$ point, which gives us the following:

```python
df['Chebyshev'] = df.apply(lambda row: max(
    abs(row['Experience'] - 10) * 0.2,
    abs(row['Ease-of-Use'] - 10) * 0.3,
    abs(row['Documentation'] - 10) * 0.1,
    abs(row['Specificity'] - 10) * 0.2,
    abs(row['Lightweightness'] - 10) * 0.1,
    abs(row['Customizability'] - 10) * 0.1
), axis=1)
df[0:5]
```

The higher the distance, the worse our option is. Thus, the priority order is the following:

1. manim
2. Unity
3. pygame
4. Processing
5. Pure Javascript

5. (2 балла) Метод анализа иерархий [Подиновский, §5.6; Петровский, §16.1-16.4; Рубчинский,
§14.1]. Если критериев у Вас очень много, то можете сократить их число до 4, чтобы Вам
пришлось меньше считать.

---

I will leave only four criteria for the sake of it: Experience, Ease-of-Use, Documentation, and Customizability.

```python
import numpy as np

criteria_matrix = np.array([
    [1, 3, 5, 2],
    [1/3, 1, 3, 1/2],
    [1/5, 1/3, 1, 1/4],
    [1/2, 2, 4, 1]
])

criteria_df = pd.DataFrame(criteria_matrix, columns=['Experience', 'Ease-of-Use', 'Documentation', 'Customizability'])
criteria_df
```

```python
normalized_matrix = criteria_matrix / criteria_matrix.sum(axis=0)

weights = normalized_matrix.mean(axis=1)
print(weights)
```

```python
data = {
    'Utility': ['Unity', 'manim', 'Processing', 'Pure Javascript', 'pygame'],
    'Experience': [4.0, 6.0, 0.1, 0.0, 2.0],
    'Ease-of-Use': [6.0, 8.0, 1.0, 0.0, 9.0],
    'Documentation': [10.0, 4.0, 5.0, 0.0, 8.0],
    'Customizability': [10.0, 5.0, 10.0, 5.0, 5.0]
}

df = pd.DataFrame(data)

df['Weighted Score'] = (df['Experience'] * weights[0] +
                        df['Ease-of-Use'] * weights[1] +
                        df['Documentation'] * weights[2] +
                        df['Customizability'] * weights[3])

df[['Utility', 'Weighted Score']]
```

```python
df['Rank'] = df['Weighted Score'].rank(ascending=False)

ranked_df = df.sort_values(by='Rank')

ranked_df[['Utility', 'Weighted Score', 'Rank']]
```
